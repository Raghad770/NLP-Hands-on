```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Natural_language_processing"

# Get HTML content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract only the main text from the article
content = soup.find_all('p')  # paragraphs only
text = "\n".join([para.get_text() for para in content])

# Save to a .txt file
with open("natural_language_processing.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("✅ Text file saved!")
```

```python
import spacy
import textacy
from textacy.extract.keyterms import sgrank

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
doc = nlp("Natural language processing and language processing are core areas in AI and NLP.")

# Get raw keyphrases
key_phrases = [term for term, _ in sgrank(doc)]
print(key_phrases)
# Sort terms by length (longest first)
key_phrases = sorted(set(key_phrases), key=len, reverse=True)

# Remove overlaps: keep a term only if it’s not a substring of an already accepted one
non_overlapping_terms = []
for term in key_phrases:
    if not any(term in bigger for bigger in non_overlapping_terms):
        non_overlapping_terms.append(term)

print("✅ Non-overlapping Key Phrases:")
print(non_overlapping_terms)
```

✅ It extracts noun chunks (noun phrases) from your document using textacy.
💡 What is a noun chunk?
A noun chunk is a group of words centered around a noun. It's usually:

A noun

Along with its modifiers (like adjectives, determiners)

📌 Examples of noun chunks:
From the sentence:

"The fast-moving car on the road was expensive."

You'd get:

"The fast-moving car"

"the road"

```python
print([chunck for chunck in textacy.extract.noun_chunks(doc)])
```
